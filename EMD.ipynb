```python
url = 'https://raw.githubusercontent.com/resendislab/POST_COVID_Metabolome_MachineLearning/main/Data/Data.csv'
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv(url)
```

```python
df = df.set_index('Folio')
```

```python
df
```

```python
label= pd.DataFrame()
label['GROUP2'] = df['GROUP2']
```

```python
df = df.iloc[:,2:]
```

```python
df
```

```python
#One vs all
import scprep
x=scprep.stats.differential_expression_by_cluster(df, label['GROUP2'], measure='emd', direction='both', gene_names=None, n_jobs=-2)
```

```python
x
```

```python
emd = pd.DataFrame([       x['CONTROL']['emd'].rename('CONTROL'),      x['COVID-19']['emd'].rename('COVID-19'),
               x['POST-COVID']['emd'].rename('POST-COVID')])
```

```python
emdlist=emd.iloc[:,:50].T
```

```python
emdlist
```

```python
import seaborn as sns
fig = sns.clustermap(emd.iloc[:,:40].T,cmap=sns.diverging_palette(250, 10, as_cmap=True),z_score=0,cbar_pos=None, figsize=(10, 12), annot=True, fmt=".2f" )
fig.savefig('EMD.png', dpi=600)
```
